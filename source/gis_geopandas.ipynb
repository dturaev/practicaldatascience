# Vector Data in Python with Geopandas

As noted in our introductory reading, GIS data comes in two flavors: vector data and raster data. 


```python
import geopandas as gpd
import numpy as np

# Load GIS data
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Drop antarctica and adjust projection
# (we'll talk about projections later!)
world = world[world.continent != "Antarctica"]
world = world.to_crs(
    "+proj=cea +lon_0=0 +lat_ts=45 +x_0=0 +y_0=0 +ellps=WGS84 +units=m +no_defs "
)
world.plot()
```
